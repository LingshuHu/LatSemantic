```python
from collections import Counter
import numpy as np
import pandas as pd
import re
import os
import string
import json
```

```python
files = os.listdir('../data_week_text/usa')
files.sort()
files
```

```python
## read metaphor and misinformation words pairs
with open('../data/all_keywords_pairs.json') as f:
  kwp = json.load(f)

## get unique key words
kw = []
for p in kwp:
    kw.append(p[0])
kw = list(set(kw))


## loop files to get frequency
for i in range(len(files)):
    with open("../data_week_text/usa/"+files[i]) as f:
        text = f.readlines()
    # obtain words frequency
    words = []
    for t in text:
        words += t.strip().split()
    c = Counter(words)
    
    frequency = []
    for w in kw:
        frequency.append(c.get(w))
    
    kw_frequency = pd.DataFrame(data = zip(kw, frequency, rep(len(words), len(kw))), 
                                columns = ["words", "frenquency", "total_words"])
    kw_frequency.to_csv("../results/usa/"+re.sub("\.txt", "", files[i])+"key_words_frequency.csv", index = False)
```
